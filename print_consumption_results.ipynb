```python
import matplotlib.pyplot as plt
import csv
```

```python
# load the results.csv file
with open('results.csv', 'r') as f:
    reader = csv.reader(f)
    data = list(reader)
    
```

```python
# scatter the data x is data[i][0] and y is data[i][1]
x = [float(row[0]) for row in data]
y = [float(row[1]) for row in data]

# compute the mean for each equal x
x_dict = {}
for i in range(len(x)):
    if x[i] in x_dict:
        x_dict[x[i]].append(y[i])
    else:
        x_dict[x[i]] = [y[i]]

x_mean = []
y_mean = []

for key in x_dict:
    x_mean.append(key)
    y_mean.append(sum(x_dict[key])/len(x_dict[key]))

# name the x axis as number of samples and name the y axis as memory consumtion in MB
plt.xlabel('Number of samples per audio clip')
plt.ylabel('Memory consumption in MB')

plt.scatter(x, y)

```
